```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
iris.data.shape
```

```python
x = iris.data
y = iris.target
x
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
single_tree=DecisionTreeClassifier(random_state=42)
single_tree.fit(x_train,y_train)
```

```python
y_pred_single = single_tree.predict(x_test)
```

```python
accuracy_single = accuracy_score(y_test,y_pred_single)
```

```python
print(f'Accuracy for Single Decision Tree :{accuracy_single}')
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bagging_model = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, random_state=42)
bagging_model.fit(x_train,y_train)
```

```python
y_pred_bagging = bagging_model.predict(x_test)
accuracy_bagging = accuracy_score(y_test,y_pred_bagging)
```

```python
print(f'Accuracy for Single Decision Tree :{accuracy_bagging}')
```

```python
from sklearn.model_selection import cross_val_predict
```

```python
cross_val_single = cross_val_predict(single_tree,x,y,cv=10)
print(f'Accuracy for the Single Tree {np.mean(cross_val_single)}')
```

```python
cross_val_bagging = cross_val_predict(bagging_model,x,y,cv=10)
print(f'Accuracy for the  {np.mean(cross_val_bagging)}')
```

```python
# So Bagging gets the best Accuracy than the decision Tree
```

```python
def plot_bootstrap_samples(x,n_samples=3):
  fig,axes=plt.subplots(n_samples,1,figsize=(10,n_samples*2))
  for i in range(n_samples):
    sample_indices=np.random.choice(range(len(x)),size=len(x),replace=True)
    sample_data=x.iloc[sample_indices,0]
    axes[i].scatter(sample_indices,sample_data,color='blue',alpha=0.5)
    axes[i].set_title(f'Bootstrap Samples {i+1}')
    axes[i].set_xlabel('Index')
    axes[i].set_ylabel('Featre 1 Value')

  plt.tight_layout()
  plt.show()

plot_bootstrap_samples(x_train)
```
